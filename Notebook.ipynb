```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import numpy as np


from src.utils.loss import Loss
from src.linear_regression.model import LinearRegression

linear_model = LinearRegression()
```

```python
feature_1 = np.random.rand(100) 
feature_2 = np.random.rand(100) 

y = 2 * feature_1 + 100 + np.random.random(100)
# + 3 * feature_2 + - 6

X_train = pd.DataFrame(
    {
    "feature1": feature_1
    #  "feature2": feature_2
    }
)  
Y_train = y 
```

```python
weights, intercepts, loss_history =  linear_model.fit(X_train, Y_train)

display(weights, intercepts, loss_history)

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.plot(loss_history)
plt.xlabel('iteration')
plt.ylabel('mse')
plt.title('Loss History')
plt.show()

predictions = linear_model.predict(X_train)

plt.figure(figsize=(8, 4))
plt.scatter(x=X_train['feature1'], y=Y_train, color='blue', label='Actual Data')
plt.scatter(x=X_train['feature1'], y=predictions, color='red', label='Predictions')
plt.xlabel('feature1')
plt.ylabel('TARGET')
plt.title('Actual vs Predictions')
plt.legend()
plt.show()


```
